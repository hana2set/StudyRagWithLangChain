```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader("../Data/The_Adventures_of_Tom_Sawyer.pdf")
document = loader.load()
document[5].page_content[:5000]
```

```python

from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(
    model="nomic-embed-text"
    ) # 

text = "   .    ?"
text_embedding = embeddings.embed_query(text)
print(text_embedding)

from langchain.vectorstores import FAISS
db = FAISS.from_documents(document, embeddings)

```

```python
# Create a vector store with a sample text
from langchain_core.vectorstores import InMemoryVectorStore

text = "LangChain is the framework for building context-aware reasoning applications"

embeddings = OllamaEmbeddings(
    model="nomic-embed-text"
    ) # 

vectorstore = InMemoryVectorStore.from_texts(
    [text],
    embedding=embeddings,
)

# Use the vectorstore as a retriever
retriever = vectorstore.as_retriever()

# Retrieve the most similar text
retrieved_documents = retriever.invoke("What is LangChain?")

# show the retrieved document's content
retrieved_documents[0].page_content
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "sentence-transformers/all-MiniLM-L6-v2")

text = "   .    ?"
text_embedding = embeddings.embed_query(text)
print(text_embedding)

from langchain.vectorstores import FAISS
db = FAISS.from_documents(document, embeddings)
```

```python
import os
os.environ["HUGGINGFACEHUB_API_TOKEN"] = "hf"  #  


from langchain.chains import RetrievalQA

from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint

hfEndpoint = HuggingFaceEndpoint(
    repo_id="MLP-KTLim/llama-3-Korean-Bllossom-8B",
    task="text-generation",
    max_new_tokens=512,
    do_sample=False,
    repetition_penalty=1.03,
) # type: ignore

llm = ChatHuggingFace(llm=hfEndpoint)

# from langchain_ollama import OllamaLLM
# llm = OllamaLLM(model="exaone3.5")

retriever = db.as_retriever()

qa = RetrievalQA.from_chain_type(
    llm=llm, 
    chain_type="stuff", 
    retriever=retriever)

query = "      ?"
result = qa.invoke({"query": query})
print(result['result'])
```
